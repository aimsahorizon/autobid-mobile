# AutoBid AI Training Notebook

2. **Body Type Classifier** - Identifies body type (Sedan/SUV/Pickup/etc.) from full car photo

````python
# Step 1: Check dataset folders
import os

def check_dataset(dataset_dir, name):
    """Print dataset summary."""
    if not os.path.exists(dataset_dir):
        print(f"ERROR: '{dataset_dir}' not found! Create the folder and add images.")
        return False
    
    classes = sorted([d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))])
    if not classes:
        print(f"ERROR: No subfolders found in '{dataset_dir}'!")
        return False
    
    print(f"\n{'='*50}")
    print(f"{name} DATASET")
    print(f"{'='*50}")
    total = 0
    for cls in classes:
        path = os.path.join(dataset_dir, cls)
        count = len([f for f in os.listdir(path) if f.lower().endswith(('.jpg','.jpeg','.png','.webp'))])
        total += count
        status = 'OK' if count >= 30 else 'LOW'
        print(f"  {cls:25s} -> {count:4d} images  [{status}]")
    print(f"{'='*50}")
    print(f"  Total: {total} images across {len(classes)} classes")
    print(f"{'='*50}")
    return True

logo_ok = check_dataset('dataset_logos', 'BRAND LOGO')
body_ok = check_dataset('dataset_body', 'BODY TYPE')
````

# PART A: Train Brand Logo Classifier
This model identifies the car brand from a close-up photo of the emblem/badge.

````python
# Step A1: Import everything we need
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV3Large
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import numpy as np
import datetime
import os

print("All imports successful!")
````

````python
# Step A2: Configuration for Brand training
BRAND_DATASET = 'dataset_logos'
IMG_SIZE = (224, 224)
BATCH_SIZE = 16

# Data generators with augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,             # Normalize pixel values to 0-1
    rotation_range=30,          # Randomly rotate up to 30 degrees
    width_shift_range=0.2,      # Randomly shift horizontally
    height_shift_range=0.2,     # Randomly shift vertically
    shear_range=0.15,           # Slight perspective distortion
    zoom_range=0.3,             # Random zoom in/out
    brightness_range=[0.7, 1.3],# Random brightness changes
    horizontal_flip=True,       # Mirror images
    fill_mode='nearest',        # Fill gaps from transforms
    validation_split=0.2        # 80% train, 20% validation
)

val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# Load images from folders
brand_train = train_datagen.flow_from_directory(
    BRAND_DATASET,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

brand_val = val_datagen.flow_from_directory(
    BRAND_DATASET,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

num_brand_classes = brand_train.num_classes
print(f"\nFound {num_brand_classes} brand classes")
print(f"Training images:   {brand_train.samples}")
print(f"Validation images: {brand_val.samples}")
````

````python
# Step A3: Save brand labels to file
brand_class_indices = brand_train.class_indices
brand_labels = {v: k for k, v in brand_class_indices.items()}

with open('brand_labels.txt', 'w') as f:
    for i in range(num_brand_classes):
        f.write(brand_labels[i] + '\n')

print("Saved brand_labels.txt:")
for i in range(num_brand_classes):
    print(f"  [{i}] {brand_labels[i]}")
````

````python
# Step A4: Build Brand Model

# Load MobileNetV3Large pretrained on ImageNet (knows general image features)
brand_base = MobileNetV3Large(
    weights='imagenet',
    include_top=False,           # Remove the original 1000-class head
    input_shape=(224, 224, 3)
)
brand_base.trainable = False     # Freeze: don't change pretrained weights yet

# Add our custom classification head
brand_model = models.Sequential([
    brand_base,
    layers.GlobalAveragePooling2D(),   # Compress features to 1D
    layers.BatchNormalization(),        # Stabilize training
    layers.Dense(128, activation='relu'),  # Learn brand-specific features
    layers.Dropout(0.3),               # Prevent overfitting
    layers.Dense(num_brand_classes, activation='softmax')  # Output: probability per brand
])

brand_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

brand_model.summary()
````

````python
# Step A5: Compute class weights (handles imbalanced classes)
from sklearn.utils.class_weight import compute_class_weight

brand_class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(brand_train.classes),
    y=brand_train.classes
)
brand_class_weights = dict(enumerate(brand_class_weights))
print("Class weights (higher = fewer images, model pays more attention):")
for idx, w in brand_class_weights.items():
    print(f"  {brand_labels[idx]:20s} -> weight {w:.3f}")
````

````python
# Step A6: PHASE 1 - Train head only (base model frozen)
# The pretrained MobileNet already knows shapes/edges/textures.
# We only train our new Dense layers to learn "which brand is this?"

brand_callbacks = [
    ModelCheckpoint('best_brand_model.keras', monitor='val_accuracy', save_best_only=True, verbose=1),
    EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7, verbose=1),
]

print("PHASE 1: Training classification head (base frozen)...")
print("This teaches the model to map MobileNet features -> brand names")
print()

brand_history1 = brand_model.fit(
    brand_train,
    validation_data=brand_val,
    epochs=25,
    class_weight=brand_class_weights,
    callbacks=brand_callbacks,
    verbose=1
)

print(f"\nPhase 1 complete!")
print(f"Best val accuracy: {max(brand_history1.history['val_accuracy'])*100:.1f}%")
````

````python
# Step A7: PHASE 2 - Fine-tune entire model
# Now we unfreeze the base model and train EVERYTHING with a very small learning rate.
# This lets the model slightly adjust its understanding of shapes/colors for car logos.

# Reload the best model from Phase 1
if os.path.exists('best_brand_model.keras'):
    brand_model = tf.keras.models.load_model('best_brand_model.keras')
    print("Loaded best model from Phase 1")

# Unfreeze the base
brand_model.layers[0].trainable = True

# IMPORTANT: Use a very small learning rate to avoid destroying pretrained knowledge
brand_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),  # 100x smaller than Phase 1
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("PHASE 2: Fine-tuning all layers...")
print("All layers unfrozen, learning rate = 0.00001")
print()

brand_history2 = brand_model.fit(
    brand_train,
    validation_data=brand_val,
    epochs=45,           # Total epochs (continues from Phase 1)
    initial_epoch=len(brand_history1.epoch),
    class_weight=brand_class_weights,
    callbacks=brand_callbacks,
    verbose=1
)

print(f"\nPhase 2 complete!")
print(f"Best val accuracy: {max(brand_history2.history['val_accuracy'])*100:.1f}%")
````

````python
# Step A8: Plot Brand training results
import matplotlib.pyplot as plt

# Combine both phases
acc = brand_history1.history['accuracy'] + brand_history2.history['accuracy']
val_acc = brand_history1.history['val_accuracy'] + brand_history2.history['val_accuracy']
loss = brand_history1.history['loss'] + brand_history2.history['loss']
val_loss = brand_history1.history['val_loss'] + brand_history2.history['val_loss']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(acc, label='Train Accuracy')
ax1.plot(val_acc, label='Val Accuracy')
ax1.axvline(x=len(brand_history1.epoch)-1, color='r', linestyle='--', label='Phase 2 start')
ax1.set_title('Brand Model - Accuracy')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)

ax2.plot(loss, label='Train Loss')
ax2.plot(val_loss, label='Val Loss')
ax2.axvline(x=len(brand_history1.epoch)-1, color='r', linestyle='--', label='Phase 2 start')
ax2.set_title('Brand Model - Loss')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.savefig('brand_training_plot.png', dpi=150)
plt.show()
print("Saved brand_training_plot.png")
````

````python
# Step A9: Export Brand model to TFLite (for Flutter)

# Load the best model
best_brand = tf.keras.models.load_model('best_brand_model.keras')

# Convert
converter = tf.lite.TFLiteConverter.from_keras_model(best_brand)
converter.optimizations = [tf.lite.Optimize.DEFAULT]  # Quantize for smaller size
tflite_brand = converter.convert()

# Save
with open('brand_model.tflite', 'wb') as f:
    f.write(tflite_brand)

size_mb = os.path.getsize('brand_model.tflite') / (1024*1024)
print(f"Saved brand_model.tflite ({size_mb:.1f} MB)")
print()
print("NEXT: Copy these 2 files to your Flutter project:")
print("  brand_model.tflite  -> assets/ai/brand_model.tflite")
print("  brand_labels.txt    -> assets/ai/brand_labels.txt")
````

# PART B: Train Body Type Classifier
This model identifies the body type (Sedan, SUV, Pickup, etc.) from a full car photo.

````python
# Step B1: Load body type dataset
BODY_DATASET = 'dataset_body'

body_train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.1,
    zoom_range=0.2,
    brightness_range=[0.7, 1.3],
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)

body_val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

body_train = body_train_datagen.flow_from_directory(
    BODY_DATASET,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

body_val = body_val_datagen.flow_from_directory(
    BODY_DATASET,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

num_body_classes = body_train.num_classes
print(f"\nFound {num_body_classes} body type classes")
print(f"Training images:   {body_train.samples}")
print(f"Validation images: {body_val.samples}")
````

````python
# Step B2: Save body labels
body_class_indices = body_train.class_indices
body_labels = {v: k for k, v in body_class_indices.items()}

with open('body_labels.txt', 'w') as f:
    for i in range(num_body_classes):
        f.write(body_labels[i] + '\n')

print("Saved body_labels.txt:")
for i in range(num_body_classes):
    print(f"  [{i}] {body_labels[i]}")
````

````python
# Step B3: Build Body Type Model

body_base = MobileNetV3Large(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)
body_base.trainable = False

body_model = models.Sequential([
    body_base,
    layers.GlobalAveragePooling2D(),
    layers.BatchNormalization(),
    layers.Dense(256, activation='relu'),   # Larger layer for body shapes (more complex)
    layers.Dropout(0.4),                    # Higher dropout
    layers.Dense(num_body_classes, activation='softmax')
])

body_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

body_model.summary()
````

````python
# Step B4: Class weights for body types
body_class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(body_train.classes),
    y=body_train.classes
)
body_class_weights = dict(enumerate(body_class_weights))
print("Body type class weights:")
for idx, w in body_class_weights.items():
    print(f"  {body_labels[idx]:15s} -> weight {w:.3f}")
````

````python
# Step B5: PHASE 1 - Train head only

body_callbacks = [
    ModelCheckpoint('best_body_model.keras', monitor='val_accuracy', save_best_only=True, verbose=1),
    EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7, verbose=1),
]

print("PHASE 1: Training body type classification head...")
print()

body_history1 = body_model.fit(
    body_train,
    validation_data=body_val,
    epochs=25,
    class_weight=body_class_weights,
    callbacks=body_callbacks,
    verbose=1
)

print(f"\nPhase 1 complete!")
print(f"Best val accuracy: {max(body_history1.history['val_accuracy'])*100:.1f}%")
````

````python
# Step B6: PHASE 2 - Fine-tune

if os.path.exists('best_body_model.keras'):
    body_model = tf.keras.models.load_model('best_body_model.keras')
    print("Loaded best model from Phase 1")

body_model.layers[0].trainable = True

body_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("PHASE 2: Fine-tuning all layers...")
print()

body_history2 = body_model.fit(
    body_train,
    validation_data=body_val,
    epochs=45,
    initial_epoch=len(body_history1.epoch),
    class_weight=body_class_weights,
    callbacks=body_callbacks,
    verbose=1
)

print(f"\nPhase 2 complete!")
print(f"Best val accuracy: {max(body_history2.history['val_accuracy'])*100:.1f}%")
````

````python
# Step B7: Plot Body Type training results

acc = body_history1.history['accuracy'] + body_history2.history['accuracy']
val_acc = body_history1.history['val_accuracy'] + body_history2.history['val_accuracy']
loss = body_history1.history['loss'] + body_history2.history['loss']
val_loss = body_history1.history['val_loss'] + body_history2.history['val_loss']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(acc, label='Train Accuracy')
ax1.plot(val_acc, label='Val Accuracy')
ax1.axvline(x=len(body_history1.epoch)-1, color='r', linestyle='--', label='Phase 2 start')
ax1.set_title('Body Type Model - Accuracy')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)

ax2.plot(loss, label='Train Loss')
ax2.plot(val_loss, label='Val Loss')
ax2.axvline(x=len(body_history1.epoch)-1, color='r', linestyle='--', label='Phase 2 start')
ax2.set_title('Body Type Model - Loss')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.savefig('body_training_plot.png', dpi=150)
plt.show()
print("Saved body_training_plot.png")
````

````python
# Step B8: Export Body model to TFLite

best_body = tf.keras.models.load_model('best_body_model.keras')

converter = tf.lite.TFLiteConverter.from_keras_model(best_body)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_body = converter.convert()

with open('body_model.tflite', 'wb') as f:
    f.write(tflite_body)

size_mb = os.path.getsize('body_model.tflite') / (1024*1024)
print(f"Saved body_model.tflite ({size_mb:.1f} MB)")
print()
print("NEXT: Copy these 2 files to your Flutter project:")
print("  body_model.tflite  -> assets/ai/body_model.tflite")
print("  body_labels.txt    -> assets/ai/body_labels.txt")
````

# Step C: Final Verification & Confusion Matrix
Test both models on the validation set and show which classes confuse the model.

````python
# Step C1: Confusion Matrix for Brand model
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Predict on validation set
brand_val.reset()
brand_preds = best_brand.predict(brand_val, verbose=1)
brand_pred_classes = np.argmax(brand_preds, axis=1)
brand_true_classes = brand_val.classes[:len(brand_pred_classes)]

# Classification report
brand_names = [brand_labels[i] for i in range(num_brand_classes)]
print("BRAND MODEL - Classification Report:")
print(classification_report(brand_true_classes, brand_pred_classes, target_names=brand_names))

# Confusion matrix heatmap
cm = confusion_matrix(brand_true_classes, brand_pred_classes)
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=brand_names, yticklabels=brand_names, ax=ax)
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
ax.set_title('Brand Model - Confusion Matrix')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('brand_confusion_matrix.png', dpi=150)
plt.show()
````

````python
# Step C2: Confusion Matrix for Body Type model

body_val.reset()
body_preds = best_body.predict(body_val, verbose=1)
body_pred_classes = np.argmax(body_preds, axis=1)
body_true_classes = body_val.classes[:len(body_pred_classes)]

body_names = [body_labels[i] for i in range(num_body_classes)]
print("BODY TYPE MODEL - Classification Report:")
print(classification_report(body_true_classes, body_pred_classes, target_names=body_names))

cm = confusion_matrix(body_true_classes, body_pred_classes)
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', xticklabels=body_names, yticklabels=body_names, ax=ax)
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
ax.set_title('Body Type Model - Confusion Matrix')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('body_confusion_matrix.png', dpi=150)
plt.show()
````
